```python
import numpy as np
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
from numpy import random

data = np.load("formationchannels.npy")
plt.hist(data, bins=100)
plt.show()
gm = []
mean = []
std = []
x = np.linspace(np.min(data), np.max(data))
x = x[:,np.newaxis]
for i in range(10):
    gm.append(GaussianMixture(n_components = i+1, random_state = 0).fit(data))
    mean.append(gm[i].means_)
    std.append(gm[i].precisions_)
    plt.hist(gm[i].score_samples(x), bins=10)
    plt.show()
```
