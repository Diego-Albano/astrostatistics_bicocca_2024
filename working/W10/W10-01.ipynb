```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
day = []
day.append(1)       #0 clear, 1 cloudy

choice = [0, 1]     #clear/cloudy
for i in range(100000):
    if day[i] == 0:
        day.append(np.random.choice(choice, p=[0.9, 0.1]))
    if day[i] == 1:
        day.append(np.random.choice(choice, p=[0.5, 0.5]))

day = np.array(day)
```

```python
weather = np.cumsum(day)/(np.arange(day.shape[0])+1)
plt.plot(weather)
```

```python
plt.hist(weather, bins=316, density=True);
print(np.mean(weather))
print(np.std(weather))
```

```python
burn = 2000
weather = weather[burn:]
plt.hist(weather, bins=316, density=True);
print(np.mean(weather))
print(np.std(weather))
```
